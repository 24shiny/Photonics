```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Ru_nk.csv')
df.head()
```

```python
df['wl'] = df['wl'].apply(lambda x: x*1000) # to nm scale
df = df.rename(columns={'wl':'wavelength [nm]'})

# add the penetration depth column
import math
df['absorption coefficient'] = (4*math.pi*df['k'])/df['wavelength [nm]']
df['penetration depth [nm]'] = 1/df['absorption coefficient']
df.head()
```

```python
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('wavelength [nm]')
ax1.set_ylabel('refractive index (n)', color=color)
ax1.plot(df['wavelength [nm]'], df['n'], '.-', color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second Axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('extinction coefficient (k)', color=color)  # we already handled the x-label with ax1
ax2.plot(df['wavelength [nm]'], df['k'], '.-', color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title('Optical Properties of Ru')
plt.show()
```

```python
# target range : 1.5 - 20 nm
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('wavelength [nm]')
ax1.set_ylabel('refractive index (n)', color=color)
ax1.plot(df['wavelength [nm]'], df['n'], '.-', color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second Axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('extinction coefficient (k)', color=color)  # we already handled the x-label with ax1
ax2.plot(df['wavelength [nm]'], df['k'], '.-', color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.xlim(0,20)
plt.title('Optical Properties of Ru in EUV')
plt.show()
```

```python
plt.plot(df['wavelength [nm]'],df['penetration depth [nm]'],'.-',color='black')
plt.xlabel('wavelength [nm]')
plt.ylabel('penetration depth [nm]')
plt.title('penetration depth of Ru')
plt.show()
```

```python
import numpy as np
plt.plot(df['wavelength [nm]'],df['penetration depth [nm]'],'.-',color='black')
plt.xlabel('wavelength [nm]')
plt.ylabel('penetration depth [nm]')
plt.title('penetration depth of Ru in EUV')
plt.xlim(0,20)
plt.xticks(np.arange(0,21,5))
plt.ylim(0,250)
plt.show()
print('mamximum penetration depth is ', df['penetration depth [nm]'].max(), 'nm')
```
