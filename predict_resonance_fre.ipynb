```python
from tabulate import tabulate
import math

def predictFreq1(x):
    return -0.207*math.log(x)+0.8734
def predictFreq2(x):
    return -0.02*x+1.062
def calculateK(x):
    return math.pow(10,4)/x
    
thickness = [0.5, 1, 1.5, 2.5, 3, 3.5, 4.5, 5, 5.5, 8, 20]
freq = [predictFreq1(i) for i in thickness]
wavenum = [calculateK(j) for j in freq]

# assign data
mydata = list(zip(thickness, freq, wavenum))

# create header
head = ["thickness [nm]", "w0_predicted [um]", "wavenumber [cm-1]"]

# display table
print(tabulate(mydata, headers=head, tablefmt="grid", showindex="always"))
```
