```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
r0_5 = pd.read_csv('Ru_FTIR/0.5nm_R_600-5000.txt', sep='\t', header=None)
t0_5 = pd.read_csv('Ru_FTIR/0.5nm_T_600-5000.txt', sep='\t', header=None)

r1 = pd.read_csv('Ru_FTIR/1nm_R_600-5000.txt', sep='\t', header=None)
t1 = pd.read_csv('Ru_FTIR/1nm_T_600-5000.txt', sep='\t', header=None)

r1_5 = pd.read_csv('Ru_FTIR/1.5nm_R_600-5000.txt', sep='\t', header=None)
t1_5 = pd.read_csv('Ru_FTIR/1.5nm_T_600-5000.txt', sep='\t', header=None)

r2 = pd.read_csv('Ru_FTIR/2nm_R_600-5000.txt', sep='\t', header=None)
t2 = pd.read_csv('Ru_FTIR/2nm_T_600-5000.txt', sep='\t', header=None)

r2_5 = pd.read_csv('Ru_FTIR/2.5nm_R_600-5000.txt', sep='\t', header=None)
t2_5 = pd.read_csv('Ru_FTIR/2.5nm_T_600-5000.txt', sep='\t', header=None)

r3 = pd.read_csv('Ru_FTIR/3nm_R_600-5000.txt', sep='\t', header=None)
t3 = pd.read_csv('Ru_FTIR/3nm_T_600-5000.txt', sep='\t', header=None)

r3_5 = pd.read_csv('Ru_FTIR/3.5nm_R_600-5000.txt', sep='\t', header=None)
t3_5 = pd.read_csv('Ru_FTIR/3.5nm_T_600-5000.txt', sep='\t', header=None)

r4_5 = pd.read_csv('Ru_FTIR/4.5nm_R_600-5000.txt', sep='\t', header=None)
t4_5 = pd.read_csv('Ru_FTIR/4.5nm_T_600-5000.txt', sep='\t', header=None)

r5 = pd.read_csv('Ru_FTIR/5nm_R_600-5000.txt', sep='\t', header=None)
t5 = pd.read_csv('Ru_FTIR/5nm_T_600-5000.txt', sep='\t', header=None)

r5_5 = pd.read_csv('Ru_FTIR/5.5nm_R_600-5000.txt', sep='\t', header=None)
t5_5 = pd.read_csv('Ru_FTIR/5.5nm_T_600-5000.txt', sep='\t', header=None)

r8 = pd.read_csv('Ru_FTIR/8nm_R_600-5000.txt', sep='\t', header=None)
t8 = pd.read_csv('Ru_FTIR/8nm_T_600-5000.txt', sep='\t', header=None)

r20 = pd.read_csv('Ru_FTIR/20nm_R_600-5000.txt', sep='\t', header=None)
t20 = pd.read_csv('Ru_FTIR/20nm_T_600-5000.txt', sep='\t', header=None)
```

```python
# merge the two

r0_5 = r0_5.rename(columns={0:'k', 1:'R'})
t0_5 = t0_5.rename(columns={0:'k', 1:'T'})
df0_5 = pd.merge(r0_5, t0_5)
df0_5['A'] = 1 - df0_5['R'] - df0_5['T']

r1 = r1.rename(columns={0:'k', 1:'R'})
t1 = t1.rename(columns={0:'k', 1:'T'})
df1 = pd.merge(r1, t1)
df1['A'] = 1 - df1['R'] - df1['T']

r1_5 = r1_5.rename(columns={0:'k', 1:'R'})
t1_5 = t1_5.rename(columns={0:'k', 1:'T'})
df1_5 = pd.merge(r1_5, t1_5)
df1_5['A'] = 1 - df1_5['R'] - df1_5['T']

r2 = r2.rename(columns={0:'k', 1:'R'})
t2 = t2.rename(columns={0:'k', 1:'T'})
df2 = pd.merge(r2, t2)
df2['A'] = 1 - df2['R'] - df2['T']

r2_5 = r2_5.rename(columns={0:'k', 1:'R'})
t2_5 = t2_5.rename(columns={0:'k', 1:'T'})
df2_5 = pd.merge(r2_5, t2_5)
df2_5['A'] = 1 - df2_5['R'] - df2_5['T']

r3 = r3.rename(columns={0:'k', 1:'R'})
t3 = t3.rename(columns={0:'k', 1:'T'})
df3 = pd.merge(r3, t3)
df3['A'] = 1 - df3['R'] - df3['T']

r3_5 = r3_5.rename(columns={0:'k', 1:'R'})
t3_5 = t3_5.rename(columns={0:'k', 1:'T'})
df3_5 = pd.merge(r3_5, t3_5)
df3_5['A'] = 1 - df3_5['R'] - df3_5['T']

r4_5 = r4_5.rename(columns={0:'k', 1:'R'})
t4_5 = t4_5.rename(columns={0:'k', 1:'T'})
df4_5 = pd.merge(r4_5, t4_5)
df4_5['A'] = 1 - df4_5['R'] - df4_5['T']

r5 = r5.rename(columns={0:'k', 1:'R'})
t5 = t5.rename(columns={0:'k', 1:'T'})
df5 = pd.merge(r5, t5)
df5['A'] = 1 - df5['R'] - df5['T']

r5_5 = r5_5.rename(columns={0:'k', 1:'R'})
t5_5 = t5_5.rename(columns={0:'k', 1:'T'})
df5_5 = pd.merge(r5_5, t5_5)
df5_5['A'] = 1 - df5_5['R'] - df5_5['T']

r8 = r8.rename(columns={0:'k', 1:'R'})
t8 = t8.rename(columns={0:'k', 1:'T'})
df8 = pd.merge(r8, t8)
df8['A'] = 1 - df8['R'] - df8['T']

r20 = r20.rename(columns={0:'k', 1:'R'})
t20 = t20.rename(columns={0:'k', 1:'T'})
df20 = pd.merge(r20, t20)
df20['A'] = 1 - df20['R'] - df20['T']

dfs = [df0_5, df1, df1_5, df2, df2_5, df3, df3_5, df4_5, df5, df5_5, df8, df20]
dfs[0].head()
```

```python
plt.plot(df5['k'], df5['R'],'.-', color='red')
plt.plot(df5['k'], df5['T'],'.-', color='blue')
plt.plot(df5['k'], df5['A'],'.-', color='black')
plt.xlabel('k [cm^(-1)]')
plt.legend(['R','T','A'], bbox_to_anchor=(1,1.02), loc='upper left')
plt.title('Experimentally Measured R & T & A of SiN+Ru_5nm')
plt.show()
```

```python
n_row = 6
n_col = 2
title = [0.5, 1, 1.5, 2, 2.5, 3, 3.5, 4.5, 5, 5.5, 8, 20]
fig, axs = plt.subplots(n_row,n_col, figsize=(10,20))
fig.tight_layout(pad=5)

for i in range(n_row):
    for j in range(n_col):
        axs[i,j].plot(dfs[2*i+j]['k'], dfs[2*i+j]['R'],'.-', color='red')
        axs[i,j].plot(dfs[2*i+j]['k'], dfs[2*i+j]['T'],'.-', color='blue')
        axs[i,j].plot(dfs[2*i+j]['k'], dfs[2*i+j]['A'],'.-', color='black')
        axs[i,j].set_xlabel('k [cm-1]')
        axs[i,j].set_ylim(0,1)
        axs[i,j].legend(['R','T','A'], bbox_to_anchor=(1,1.02), loc='upper left')
        axs[i,j].set_title(f'R & T & A of SiN+Ru_{title[2*i+j]}nm')
```

```python
from dash import Dash, dcc, html, Input, Output
import plotly.graph_objects as go
import plotly.express as px

app = Dash(__name__)


app.layout = html.Div([
    html.H4('R & T & A of Si+Ru films'),
    html.P("Select thickness of Ru :"),
    dcc.Dropdown(
        id="dropdown",
        options=['0.5 nm','1 nm','1.5 nm','2 nm','2.5 nm','3 nm','3.5 nm','4.5 nm','5 nm','5.5 nm','8 nm','20 nm'],
        value='0.5 nm',
        clearable=False,
    ),
    dcc.Graph(id="graph"),
    dcc.Slider(
        id='slider-position1', 
        min=600, max=1400, value=650, step=0.1,
        marks={600: '600', 1400: '1400'}
    ),
    dcc.Slider(
    id='slider-position2', 
    min=600, max=1400, value=1400, step=0.1,
    marks={600: '600', 1400: '1400'}
    )   
])


@app.callback(
    Output("graph", "figure"), 
    Input("dropdown", "value"),
    Input("slider-position1", "value"), Input("slider-position2", "value"))
def display_graph(opt, pos1, pos2):
    # map an option to an index of a df concerned
    options = ['0.5 nm','1 nm','1.5 nm','2 nm','2.5 nm','3 nm','3.5 nm','4.5 nm','5 nm','5.5 nm','8 nm','20 nm']
    df_target = dfs[options.index(opt)]
    fig = go.Figure(
        data = [go.Scatter(x=df_target['k'].values, y=df_target['R'], line_color='red', name='R'),
                # go.Scatter(x=[df_target[df_target['R']==df_target['R'][:1000].min()]['k'], 
                #               df_target[df_target['R']==df_target['R'][:1000].max()]['k']],
                #           y=[df_target['R'].min(), df_target['R'].max()], mode='markers', marker=dict(color='black', size=20)),
               go.Scatter(x=df_target['k'].values, y=df_target['T'], line_color='blue', name='T'),
               go.Scatter(x=df_target['k'].values, y=df_target['A'], line_color='black', name='A')])
    fig.update_layout(plot_bgcolor="white", xaxis_range=[600,5000], yaxis_range=[0,1], xaxis_title='wavenumber [cm-1]', margin=dict(l=20, r=20, t=20, b=20),)
    fig.update_xaxes(
    mirror=True,
    ticks='outside',
    showline=True,
    linecolor='black',
    gridcolor='lightgrey'
    )
    fig.update_yaxes(
        mirror=True,
        ticks='outside',
        showline=True,
        linecolor='black',
        gridcolor='lightgrey'
    )
    fig.add_vline(x=pos1, line_width=3, line_color="green", annotation_text=' L')
    fig.add_vline(x=pos2, line_width=3, line_color="green", annotation_text=' R')
    return fig


app.run_server(debug=True)
```
