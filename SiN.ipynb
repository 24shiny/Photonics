```python
import pandas as pd
df = pd.read_excel('SiN_RT.xlsx', sheet_name='Sheet1')
df.head()
```

```python
import math
df['A'] = 1 - df['R'] - df['T']
df['nm'] = df['k'].apply(lambda x: math.pow(10,7)/x)
df['um'] = df['nm'].apply(lambda x:x/1000)
df.head()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
plt.plot(df['k'], df['R'],'.-', color='red')
plt.plot(df['k'], df['T'],'.-', color='blue')
plt.plot(df['k'], df['A'],'.-', color='black')
plt.xlabel('wavenumber [cm^(-1)]')
plt.legend(['R','T','A'], bbox_to_anchor=(1,1.02), loc='upper left')
plt.title('Experimentally Measured R & T & A of SiN')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot(df['k'], df['R'],'.-', color='red')
plt.plot(df['k'], df['T'],'.-', color='blue')
plt.plot(df['k'], df['A'],'.-', color='black')
plt.xlabel('wavenumber [cm^(-1)]')
plt.xlim(600,1500)
plt.legend(['R','T','A'], bbox_to_anchor=(1,1.02), loc='upper left')
plt.axvline(x = 844.77377, color = 'black', linestyle=':', label = 'axvline - full height')
plt.axhline(y = 0.5, color = 'black', linestyle=':', label = 'axvline - full height')
plt.text(860, 0.98, 'A_max at 844.77', fontsize = 10)
plt.axvspan(850, 1200, color='yellow', alpha=0.2, lw=0)
plt.text(860, 0.9, 'w_p', fontsize = 10)
plt.title('Experimentally Measured R & T & A of SiN')
plt.show()
```
